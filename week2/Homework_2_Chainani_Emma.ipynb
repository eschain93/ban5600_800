### 1: Guassian Random Numbers

```python
# 1: Guassian random numbers

import random
import math

u = random.random() # decimal from 0 to less than 1
v = random.random() # decimal from 0 to less than 1

z = math.sin(2 * math.pi * v) * (-2 * math.log(u)) ** 0.5

print("u =", u)
print("v =", v)
print("z =", z)
```

```python
# 1a: Import random; random.uniform

import random

u = random.uniform(1, 10) #decimal between 1 and 10
v = random.uniform(1, 10) #decimal between 1 and 10

z = u*v

print("u =", u)
print("v =", v)
print("z =", z)
```

```python
# 1a: Import random; random.radint

import random

u = random.randint(1, 10) # whole number from 1 to 10, inclusive
v = random.randint(1, 10) # whole number from 1 to 10, inclusive

z = u*v

print("u =", u)
print("v =", v)
print("z =", z)
```

```python
# 1a: Import random; random.radint

import random

u = random.randrange(1, 10) # whole number from 1 to 9
v = random.randrange(1, 10) # whole number from 1 to 9

z = u*v

print("u =", u)
print("v =", v)
print("z =", z)
```

### 4: Richter Scale

```python
# 4 Print the earthquake magnitude using nested if statements

magnitude = float(input("Enter earthquake magnitude: "))

if magnitude < 2.0:
    print("A magnitude", magnitude, "earthquake is considered to be a micro earthquake.")
elif magnitude < 3.0:
    print("A magnitude", magnitude, "earthquake is considered to be a very minor earthquake.")
elif magnitude < 4.0:
    print("A magnitude", magnitude, "earthquake is considered to be a minor earthquake.")
elif magnitude < 5.0:
    print("A magnitude", magnitude, "earthquake is considered to be a light earthquake.")
elif magnitude < 6.0:
    print("A magnitude", magnitude, "earthquake is considered to be a moderate earthquake.")
elif magnitude < 7.0:
    print("A magnitude", magnitude, "earthquake is considered to be a strong earthquake.")
elif magnitude < 8.0:
    print("A magnitude", magnitude, "earthquake is considered to be a major earthquake.")
elif magnitude < 10.0:
    print("A magnitude", magnitude, "earthquake is considered to be a great earthquake.")
else:
    print("A magnitude", magnitude, "earthquake is considered to be a meteoric earthquake.")
```

```python
# 4a Print the earthquake magnitude using nested if statements and variables

magnitude = float(input("Enter earthquake magnitude: "))

if magnitude < 2.0:
    descriptor = "micro"
elif magnitude < 3.0:
    descriptor = "very minor"
elif magnitude < 4.0:
    descriptor = "minor"
elif magnitude < 5.0:
    descriptor = "light"
elif magnitude < 6.0:
    descriptor = "moderate"
elif magnitude < 7.0:
    descriptor = "strong"
elif magnitude < 8.0:
    descriptor = "major"
elif magnitude < 10.0:
    descriptor = "great"
else:
    descriptor = "meteoric"

print("A magnitude", magnitude, "earthquake is considered to be a", descriptor, "earthquake.")
```

### 6: Print Sentence

```python
# 6 Print uppercase phrase, count of letter 't', and split by space

s = 'I ate it.'
print(s.upper()) #.upper() converts the string to uppercase
print(s.count('t')) #.count('t') counts how many times the character 't' appears
w = s.split() #.split() splits the string into a list by spaces
print(w)
```

```python
# 6a Print lowercase phrase, count of letter 't', and split by space

s = "My name is Emma Chainani, and I'm an MBA student enrolled at William Paterson University."
print(s.lower()) #.lower() converts the string to lowercase
print(s.count('i')) #.count('i') counts how many times the character 'i' appears
w = s.split(",") #.split(",") splits the string into a list by comma
print(w)
```
