### 1: Guassian Random Numbers

```python
# 1: Guassian random numbers
# Solve the Box-Muller formula 𝑍 = 𝑠𝑖𝑛(2 𝜋 𝑣)(−2 𝑙𝑛 𝑢)^1/2
# Use the random and math libraries

import random
import math

u = random.random() # decimal from 0 to less than 1
v = random.random() # decimal from 0 to less than 1

z = math.sin(2 * math.pi * v) * (-2 * math.log(u)) ** 0.5

print("u =", u)
print("v =", v)
print("z =", z)
```

### 3. The Journey: Typing apple.com

*If the DNS "Phonebook" was hacked and someone changed the IP address for your bank's website to point to a fake server, how would your browser know the difference?*

When someone types their bank’s website into a browser, several steps happen before the page appears. First, DNS translates the bank’s domain name into an IP address. Next, the browser and server perform a handshake to establish a connection. Then, the browser sends an HTTP request asking for the webpage, and the server responds with the code needed to load it. Finally, the browser renders that code into the website the user sees.

If the DNS “phonebook” were hacked and the bank’s website pointed to a fake server, the main way a browser could  detect the difference is by checking the site’s security certificate. If the certificate did not match the bank’s real domain or was not issued by a trusted source, the browser would usually warn the user that the connection is not secure.

Other warning signs that would signify the DNS was hacked is the site would not render as its typical landing page. More pop up advertisements may be present and the URL in the address bar may be changed to something slightly different. For example, if https://www.bankofamerica.com/ were hacked, the final URL might redirect to: https://www.bankofamerica-account.com/
