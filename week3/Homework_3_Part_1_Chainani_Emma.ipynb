##1 [5pts] Write program with while loop which finds the followings for the given values between 150 to 1000.

a) How many odd numbers are between 150 and 1000 (include 150 and 1000)

b) Find the sum of odd numbers which are between 150 and 1000 (include 150 and 1000)

c) Find the multiplication of odd numbers which are between 150 and 1000 (include 150
and 1000)

d) Find the sum of square of odd numbers which are between 150 and 1000 (include 150
and 1000)

```python
#1 Write program with while loop which finds the followings for the given values between 150 to 1000.

current_number = 150
count_odd = 0         # Initializes a counter for odd numbers
sum_odd = 0           # Initializes a variable to accumulate the sum of odd numbers
product_odd = 1       # Initializes a variable to accumulate the product of odd numbers (starting with 1 to avoid a 0 product)
sum_sq_odd = 0        # Initializes a variable to accumulate the sum of squares of odd numbers

while current_number <= 1000: # Loop through numbers starting at 150 up to and including 1000
    if current_number % 2 != 0: # Check if the current number is odd (remainder is not 0 when divided by 2)
        count_odd += 1          # Increment the odd number counter
        sum_odd += current_number # Add the current odd number to the sum
        product_odd *= current_number # Multiply the current odd number into the product
        sum_sq_odd += (current_number ** 2) # Add the square of the current odd number to the sum of squares
    current_number += 1 # Move to the next number in the sequence

print(f"a) There are {count_odd} odd numbers between 150 and 1000")
print(f"b) The sum of odd numbers between 150 and 1000 is {sum_odd}")
print(f"c) The product of of odd numbers between 150 and 1000 is {product_odd}")
print(f"d) The sum of square of odd numbers between 150 and 1000 is {sum_sq_odd}")
```

##2 [5pts] Write program with for loop which finds the followings for the given values between 150 to 1000.

a) How many odd numbers are between 150 and 1000 (include 150 and 1000)

b) Find the sum of odd numbers which are between 150 and 1000 (include 150 and 1000)

c) Find the multiplication of odd numbers which are between 150 and 1000 (include 150
and 1000)

d) Find the sum of square of odd numbers which are between 150 and 1000 (include 150
and 1000)

```python
#2 Write program with for loop which finds the followings for the given values between 150 to 1000.

count_odd = 0
sum_odd = 0
product_odd = 1
sum_sq_odd = 0

# range(150, 1001) starts at 150 and stops before 1001 (includes 1000)
for current_number in range(150, 1001):
    if current_number % 2 != 0: # Check if the number is odd
        count_odd += 1          # Increment the counter
        sum_odd += current_number # Add to the sum
        product_odd *= current_number # Multiply to the product
        sum_sq_odd += (current_number ** 2) # Add square to the sum of squares

print(f"a) There are {count_odd} odd numbers between 150 and 1000")
print(f"b) The sum of odd numbers between 150 and 1000 is {sum_odd}")
print(f"c) The product of of odd numbers between 150 and 1000 is {product_odd}")
print(f"d) The sum of square of odd numbers between 150 and 1000 is {sum_sq_odd}")
```

## 3 [1pts] Write a program so that for given M, it computes the sum

\[
s=\sum_{k=1}^{M}\frac{1}{k}
\]

using a `while` loop.

```python
#3 Write a program so that for given M, it computes the sum [ s=\sum_{k=1}^{M}\frac{1}{k} ]

M = int(input("Enter M: "))   # get the value of M

s = 0                         # start sum at 0
k = 1                         # start k at 1

while k <= M:                 # keep looping until k reaches M
    s = s + 1/k               # add 1/k to the sum
    k = k + 1                 # increase k by 1

print("The sum is:", s)       # print final result
```

```python
#3a Write a program so that for given M, it computes the sum [ s=\sum_{k=1}^{M}\frac{1}{k} ]

M = int(input("Enter M: "))  # ask the user to enter a value for M

s = 0  # start the sum at 0
k = 1  # start counting at 1

while k <= M:  # keep looping as long as k is less than or equal to M
    print(f"Adding 1/{k}")  # show which fraction is being added
    s = s + 1/k  # add the current fraction 1/k to the sum
    print("k =", k, "current sum =", round(s, 2))  # show the running total
    k = k + 1  # increase k by 1

print(f"Final sum: {s:.2f}")  # print the final sum
```

##4 [1pts] Write a program which repeatedly reads numbers until the user enters “done”.

Once “done” is
entered, print out the total, count, and average of the numbers. If the user enters anything other than a
number, detect their mistake using try and except and print an error message and skip to the next number.

```python
#4 Write a program to calculate total, count, and average of numbers, handling invalid input.

total = 0     # Initialize total sum of numbers
count = 0     # Initialize count of valid numbers
average = 0   # Initialize average

while True:   # Loop indefinitely until 'done' is entered
    user_input = input("Enter a number: ") # Prompt user for input

    if user_input.lower() == 'done': # Check if user wants to exit
        break # Exit the loop
    try:
        number = float(user_input) # Attempt to convert input to a float
        total += number            # Add number to total
        count += 1                 # Increment count
    except ValueError: # Catch error if input is not a valid number
        print("Invalid input") # Inform user of invalid input

# Calculate average if at least one number was entered
if count > 0:
    average = total / count

# Print the results
print(f"Total: {total}, Count: {count}, Average: {average}")
```

```python
#4a Refined number processing program with better error handling and multiple exit commands.

total = 0     # Initialize total sum of numbers
count = 0     # Initialize count of valid numbers
numbers_list = [] # Initialize a list to store valid numbers (FIXED: added this line)

print("Enter numbers. Type 'done', 'quit', or 'exit' to finish.")

while True:
    user_input = input("Enter a number: ").lower().strip() # Get input, convert to lowercase, remove whitespace

    # Check for multiple exit commands
    if user_input in ['done', 'quit', 'exit']:
        break # Exit the loop
    try:
        number = float(user_input) # Attempt to convert input to a float
        total += number            # Add number to total
        count += 1                 # Increment count
        numbers_list.append(number) # Store the valid number in the list
    except ValueError: # Catch error if input is not a valid number
        print(f"Invalid input: '{user_input}'. Please enter a valid number or an exit command.")

print("\n--- Results ---")
# Display results if valid numbers were entered
if count > 0:
    average = total / count
    print(f"Total: {total}")
    print(f"Count: {count}")
    print(f"Average: {average}")
else:
    print("No valid numbers were entered.")
```

## 5 [1pts] Write another program that prompts for a list of numbers as above (question 4) and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
#5 Write a program to find the maximum and minimum of a list of numbers.

numbers_list = [] # Initialize an empty list to store valid numbers

print("Enter numbers. Type 'done', 'quit', or 'exit' to finish.")

while True:
    user_input = input("Enter a number: ").lower().strip() # Prompt user for input

    # Check if the user wants to exit
    if user_input in ['done', 'quit', 'exit']:
        break # Exit the loop
    try:
        number = float(user_input) # Attempt to convert input to a float
        numbers_list.append(number) # Add the valid number to the list
    except ValueError: # Catch error if input is not a valid number
        print(f"Invalid input: '{user_input}'. Please enter a valid number or an exit command.")

print("\n--- Results ---")
# Check if any valid numbers were entered to find min/max
if numbers_list:
    print(f"Minimum number: {min(numbers_list)}") # Print the minimum number from the list
    print(f"Maximum number: {max(numbers_list)}") # Print the maximum number from the list
else:
    print("No valid numbers were entered to find min/max.")
```
